```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/donations_2016_2020_2024.csv')
df['date'] = pd.to_datetime(df['date'])
df['week'] = df['date'].dt.isocalendar().week
weekly = df.groupby(['year', 'week'])['amount'].sum().reset_index()
pivot = weekly.pivot(index='year', columns='week', values='amount').fillna(0)
plt.figure(figsize=(12, 4))
sns.heatmap(pivot, cmap='Reds', annot=True, fmt='.0f')
plt.title('Chinese-Linked Donations by Week (2016â€“2024)')
plt.show()
```
